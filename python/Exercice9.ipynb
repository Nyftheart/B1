```
def DoubleIndice():
  R = []
  i = 0
  while i < 10:
    j = 0
    T = []
    while j < 10:
      T.append(i+j)
      j = j + 1
    i = i + 1
    R.append(T)
  return R

L = DoubleIndice()
print (L)
```

```
from random import *

def BidiRandomList(T1,T2):
  R = [] 
  i = 0 
  while i < T1:
    j = 0
    T = []
    while j < T2:
      T.append(randrange(-100,100))
      j = j + 1
    i=i+1
    R.append(T)
  return R
```

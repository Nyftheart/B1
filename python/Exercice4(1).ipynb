```
def TableDe5():
  chiffre = 5
  compteur = 0
  while compteur<10:
    Result = compteur * chiffre
    print(chiffre,"x",compteur,"=",Result)
    compteur = compteur + 1
    
TableDe5()
```

```
def TableDeX(x):
  compteur = 0
  while compteur<10:
    Result = compteur * x
    print(x,"x",compteur,"=",Result)
    compteur = compteur + 1
    
i = int(input())
TableDeX(i)
```

```
def Min (a,b):
  if a > b:
    return b
  else:
    return a
  
def Max (a,b):
  if a < b:
    return b
  else:
    return a  

def MinOrMax(a,b):
  choix = input()
  if choix == "min":
    return Min(a,b)
  elif choix == "max":
    return Max(a,b)
  

a = MinOrMax(10,5)
print (a)
```

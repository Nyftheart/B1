```
def Premier(x):
  c = 2
  IsPremier = True
  while c < x:
    if x % c == 0 :
      IsPremier = False
    c = c + 1
  return IsPremier
    
```

```
def Power(a,b):
  if b == 0:
    return 1
  else:
    return a * Power(a,b-1)

p = Power(2,4)
print (p)
```

```
def Facto(x):
  if x == 0:
    return 1
  else:
    return x * Facto(x-1)

f = Facto(5)
print(f)
```

```
def Mini (a,b) :
  if a < b:
    return a
  else :
    return b
 
def Max(a,b) :
  if a > b:
    return a
  else :
    return b

def ChoixMinMax(a,b):
  print("Ecrire min pour le minimum et max pour la maximum")
  choix = input() 
  if choix == "min":
    return Mini(a,b)
  elif choix == "max" : 
    return Max(a,b)
  
  
n1 = int(input())
n2 = int(input())
m = ChoixMinMax(n1,n2)
print(m)
```

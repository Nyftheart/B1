```
def puissance(a,b):
  comp = 0
  resultat = 1
  while comp<b:
    resultat = resultat * a
    comp = comp + 1
  return resultat

one = puissance(1,1)
two = puissance(2,2)
```

```
def SommePuissance(limite):
  comp = 1
  resultat = 0
  while comp <= limite:
    resultat = resultat + puissance(comp,comp)
    comp = comp + 1
  return resultat

r = SommePuissance(3)
print(r)
```

```
def Premier(p):
  comp = 2
  IsPremier = True
  
  while comp < p : 
    if p%comp == 0:
      IsPremier=False
    comp= comp+1
    
  return IsPremier
      
if Premier(15)==True:
  print("15 est premier")
else:
  print("15 n'est pas premier")
```

```
def LesPremiers(lim):
  comp = 1
  while comp < lim:
    if Premier(comp) == True:
      print(comp)
    comp = comp + 1
    
LesPremiers(1000)
```

```
n = int(input('Entrez un entier [1 .. 4000[ : '))

s = "" # Chaîne résultante

while n >= 1000:
    s += "M"
    n -= 1000
if n >= 900:
    s += "CM"
    n -= 900
if n >= 500:
    s += "D"
    n -= 500
if n >= 400:
    s += "CD"
    n -= 400

while n >= 100:
    s += "C"
    n -= 100
if n >= 90:
    s += "XC"
    n -= 90
if n >= 50:
    s += "L"
    n -= 50
if n >= 40:
    s += "XL"
    n -= 40
while n >= 10:
    s += "X"
    n -= 10
if n >= 9:
    s += "IX"
    n -= 9
if n >= 5:
    s += "V"
    n -= 5
if n >= 4:
    s += "IV"
    n -= 4
while n >= 1:
    s += "I"
    n -= 1

print("En romain :", s)
```
